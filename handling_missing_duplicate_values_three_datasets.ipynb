#Handling missing and duplicate values in three datasets

```python
import pandas as pd

df=pd.read_csv('Data.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
df['Salary'].fillna(df['Salary'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
duplicate_rows = df.duplicated().sum()
print(f'Number of duplicate rows: {duplicate_rows}')
```

```python
import pandas as pd

df2=pd.read_csv('Employee.csv')
df2.head()
```

```python
df2.isnull().sum()
```

```python
df2['Age'].fillna(df2['Age'].mean(),inplace=True)
```

```python
df2['Salary'].fillna(df2['Salary'].mean(),inplace=True)
```

```python
# Calculate the mode of the 'Place' column
mode_value_place = df2['Place'].mode()[0]
df2['Place'].fillna(mode_value_place,inplace=True)
```

```python
# Calculate the mode of the 'Company' column
mode_value_company = df2['Company'].mode()[0]
df2['Company'].fillna(mode_value_company,inplace=True)
```

```python
df2.isnull().sum()
```

```python
duplicate_rows = df2.duplicated().sum()
print(f'Number of duplicate rows: {duplicate_rows}')
```

```python
df2.drop_duplicates(inplace=True)
print(f'Number of duplicate rows: {df2.duplicated().sum()}')
```
